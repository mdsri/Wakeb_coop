```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, make_scorer
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

# Assuming 'data' is your original DataFrame

# 1. Copy dataset to avoid modification
df = data.copy()

# 3. Encode categorical variables
for col in ['Gender', 'Ethnicity', 'EducationLevel']:
    df[col] = LabelEncoder().fit_transform(df[col])

# 4. Select top important features
top_features = [
    'DustExposure', 'PollenExposure', 'DietQuality', 'SleepQuality',
    'BMI', 'PollutionExposure', 'LungFunctionFVC', 'PhysicalActivity', 'Age'
]

X_top = df[top_features]
y = df['LungFunctionFEV1']

# 5. Scale features
scaler = StandardScaler()
X_top_scaled = pd.DataFrame(scaler.fit_transform(X_top), columns=top_features)

# 6. Train/test split (for final test evaluation)
X_train, X_test, y_train, y_test = train_test_split(X_top_scaled, y, test_size=0.2, random_state=42)

# 7. Initialize model
model = xgb.XGBRegressor(random_state=42)

# 8. Perform K-Fold cross-validation on training data
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Define scoring for cross_val_score (use R² here)
cv_r2_scores = cross_val_score(model, X_train, y_train, cv=kf, scoring='r2')

print("K-Fold Cross-validation R² scores:", cv_r2_scores)
print(f"Mean CV R²: {cv_r2_scores.mean():.3f} ± {cv_r2_scores.std():.3f}")

# 9. Train model on full training data
model.fit(X_train, y_train)

# 10. Predict on test set
preds = model.predict(X_test)

# 11. Evaluate on test set
mae = mean_absolute_error(y_test, preds)
mse = mean_squared_error(y_test, preds)
r2 = r2_score(y_test, preds)

print(f"\nTest Set Performance:")
print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")

# 12. Plot predicted vs actual for test set
plt.figure(figsize=(6,5))
sns.scatterplot(x=y_test, y=preds, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual LungFunctionFEV1")
plt.ylabel("Predicted LungFunctionFEV1")
plt.title("XGBoost Regression - Top Features (Test Set)")
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Prepare data (reuse X_top_scaled and y from previous)
X_train, X_test, y_train, y_test = train_test_split(X_top_scaled, y, test_size=0.2, random_state=42)

models = {
    'Ridge Regression': Ridge(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    print(f"{name}:")
    print(f"  MAE: {mean_absolute_error(y_test, preds):.3f}")
    print(f"  MSE: {mean_squared_error(y_test, preds):.3f}")
    print(f"  R²: {r2_score(y_test, preds):.3f}\n")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb

# Assuming your original dataset is loaded into `data`
df = data.copy()

# 2. Encode categorical variables
for col in ['Gender', 'Ethnicity', 'EducationLevel']:
    df[col] = LabelEncoder().fit_transform(df[col])

# 3. Create grouped/generalized features
df['Lifestyle_Score'] = df[['BMI', 'Smoking', 'PhysicalActivity', 'DietQuality', 'SleepQuality']].mean(axis=1)
df['Environmental_Exposure'] = df[['PollutionExposure', 'PollenExposure', 'DustExposure']].sum(axis=1)
df['Medical_History_Score'] = df[['FamilyHistoryAsthma', 'HistoryOfAllergies', 'Eczema', 'HayFever', 'GastroesophagealReflux', 'PetAllergy']].sum(axis=1)
df['Symptoms_Score'] = df[['Wheezing', 'ShortnessOfBreath', 'ChestTightness', 'Coughing', 'NighttimeSymptoms', 'ExerciseInduced']].sum(axis=1)

# 4. Define features and target
feature_groups = [
    'Age','Lifestyle_Score', 'Environmental_Exposure', 'Medical_History_Score', 'Symptoms_Score',
    'LungFunctionFVC'
]

X = df[feature_groups]
y = df['LungFunctionFEV1']

# 5. Scale continuous features (optional but recommended)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=feature_groups)

# 6. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 7. Train XGBoost regressor
model = xgb.XGBRegressor(random_state=42)
model.fit(X_train, y_train)

# 8. Predict & evaluate
preds = model.predict(X_test)
mae = mean_absolute_error(y_test, preds)
mse = mean_squared_error(y_test, preds)
r2 = r2_score(y_test, preds)

print(f"XGBoost with Grouped Features:\nMAE: {mae:.3f}\nMSE: {mse:.3f}\nR²: {r2:.3f}")

# 9. Plot predicted vs actual
plt.figure(figsize=(6, 5))
sns.scatterplot(x=y_test, y=preds, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual LungFunctionFEV1")
plt.ylabel("Predicted LungFunctionFEV1")
plt.title("XGBoost Regression - Grouped Features")
plt.tight_layout()
plt.show()
```
