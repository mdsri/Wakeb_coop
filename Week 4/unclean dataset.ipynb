```python
import pandas as pd
import random
import numpy as np

def unclean_dataset(df, corruption_rate=0.3, seed=42):
    random.seed(seed)
    np.random.seed(seed)
    
    def dirty_date(date_str):
        try:
            if random.random() < 0.5:
                return date_str.replace('/', '-')  # e.g., 1-31-2019
            elif random.random() < 0.5:
                return date_str[::-1]  # reverse string
            else:
                return '2019/9/3O'  # typo example
        except:
            return date_str

    def dirty_text(text):
        if pd.isna(text): return text
        text = str(text)
        if random.random() < 0.5:
            text = text.lower()
        if random.random() < 0.3:
            text = text.replace('o', '0').replace('O', '0')
        if random.random() < 0.3:
            text = text.replace('i', '1').replace('e', '3')
        if random.random() < 0.3:
            text += random.choice(['!', '@', '...', '?'])
        return text

    def dirty_flag(flag):
        if flag not in ['Y', 'N', 'y', 'n']: return flag
        return random.choice(['Y', 'N', '1', '0', 'yes', 'no', '-', '??', 'n/a'])

    def dirty_gender(g):
        return random.choice(['M', 'F', 'm/f', 'unknown', g, np.nan])

    def dirty_age(age):
        return random.choice(['unkwn', '<18', '>55', 'twenty', '36–55', age, np.nan])

    def dirty_country_code(code):
        return str(code).lower().replace('u', 'v').replace('a', '@') + random.choice(['', '?'])

    def dirty_country_name(name):
        return str(name).lower().replace(' ', '') + random.choice(['??', '...', '', '!!'])

    def dirty_device_count(val):
        if random.random() < 0.3:
            return random.choice(['one', 'five', 'twelve units', 'NaN', np.nan])
        return val

    df_dirty = df.copy()
    for idx, row in df_dirty.iterrows():
        for col in df_dirty.columns:
            if random.random() > corruption_rate:
                continue  # Skip corruption on this field

            if col == 'CAL_DT':
                df_dirty.at[idx, col] = dirty_date(row[col])
            elif col in ['MODEL_NAME', 'BRAND_FULL_NAME', 'VENDOR_NAME', 'OS_NAME', 'DEVICE_TYPE']:
                df_dirty.at[idx, col] = dirty_text(row[col])
            elif col.endswith('_FLG') and col != 'DUAL_SIM_FLG':
                df_dirty.at[idx, col] = dirty_flag(row[col])
            elif col == 'GENDER_TYPE_CD':
                df_dirty.at[idx, col] = dirty_gender(row[col])
            elif col == 'AGE_B':
                df_dirty.at[idx, col] = dirty_age(row[col])
            elif col == 'NATIONALITY_CD':
                df_dirty.at[idx, col] = dirty_country_code(row[col])
            elif col == 'NATIONALITY_NAME':
                df_dirty.at[idx, col] = dirty_country_name(row[col])
            elif col == 'DEVICE_COUNT':
                df_dirty.at[idx, col] = dirty_device_count(row[col])
            # DUAL_SIM_FLG remains clean

    return df_dirty
```

```python
import pandas as pd
import random
import numpy as np
import re

def reversible_noise(text, strength=0.3):
    """
    Add reversible noise to string:
    - Randomly change case
    - Insert symbols like '.', ',', '-', '_', '@', '#', '$', '%'
    - Insert random spaces
    """
    if pd.isna(text):
        return text
    text = str(text)
    noisy = ""
    for c in text:
        r = random.random()
        if r < strength / 2:
            # Random case change
            c = c.upper() if random.random() < 0.5 else c.lower()
            noisy += c
        elif r < strength:
            # Append random symbol after character
            noisy += c + random.choice(['.', ',', '_', '@', '#', '$', '%', ' '])
        else:
            noisy += c
    return noisy

def reversible_dirty_dataset(df, corruption_rate=0.4, strength=0.3, seed=42):
    random.seed(seed)
    np.random.seed(seed)

    df_dirty = df.copy()

    for idx, row in df_dirty.iterrows():
        for col in df_dirty.columns:
            if col == 'DUAL_SIM_FLG':
                continue  # keep original, no noise here

            if random.random() > corruption_rate:
                continue  # skip some fields for partial corruption

            val = row[col]
            # Only dirty strings or numbers (convert numbers to str)
            if isinstance(val, (str, int, float)):
                dirty_val = reversible_noise(val, strength=strength)
                df_dirty.at[idx, col] = dirty_val

    return df_dirty

def simple_clean_string(text):
    """
    Clean the reversible noise by:
    - Removing all symbols except letters, digits, and spaces
    - Lowercasing
    - Stripping spaces
    """
    if pd.isna(text):
        return text
    text = str(text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text)  # collapse multiple spaces
    return text.strip().lower()

def clean_reversible_dataset(df_dirty, columns=None):
    df_cleaned = df_dirty.copy()
    cols = columns if columns else df_dirty.columns

    for col in cols:
        if col == 'DUAL_SIM_FLG':
            continue  # keep original as is
        df_cleaned[col] = df_cleaned[col].apply(simple_clean_string)

    return df_cleaned

# Example usage:
if __name__ == "__main__":
    # Load your clean dataset (replace with your actual file)
    

    # Apply reversible noise
    df_dirty = reversible_dirty_dataset(df, corruption_rate=0.5, strength=0.4)


    print("Done! Dirty and restored datasets created.")
```
